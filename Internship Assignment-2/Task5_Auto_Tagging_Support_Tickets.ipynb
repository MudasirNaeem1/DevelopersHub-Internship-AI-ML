```python
!pip install transformers --quiet
from transformers import pipeline

# Zero-shot classifier
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

tickets = [
    "My internet connection is down since morning.",
    "I want a refund for the last month’s bill.",
    "Unable to login to my account after password reset.",
    "System is very slow and crashes frequently."
]

candidate_labels = ["Billing",
                    "Technical Support",
                    "Account",
                    "Performance"]

for t in tickets:
    result = classifier(t, candidate_labels, multi_label=True)
    print("\nTicket:", t)
    print("Top Tags:", result["labels"][:3])
```
